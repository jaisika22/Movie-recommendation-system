**PRE PROCESSING**

```python
# list of attributes from 23 to recommend a movie
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies = movies[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
movies.shape
```

```python
movies.isnull().sum()
```

```python
# just incase any column had any null values we will use :
movies.dropna(inplace = True)
```

*working on the columns to extract the info.*

```python
movies.iloc[0].genres
```

```python
def convert(input):
  L=[]
  for i in ast.literal_eval(input):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies.iloc[0].cast
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter!= 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
movies.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features = 5000, stop_words= 'english')
```

```python
vectors= cv.fit_transform(new['tags']).toarray()   ## converting it into a numpy array
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()

def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new['tags']= new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features = 5000, stop_words= 'english')
```

```python
vectors= cv.fit_transform(new['tags']).toarray()   ## converting it into a numpy array
```

```python
cv.get_feature_names_out()  # these are the 5000 words that r maximum used
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    # Check if the movie is in the DataFrame after converting both to lowercase and stripping whitespaces
    movie_lower = movie.lower().strip()
    if movie_lower in new['title'].str.lower().str.strip().values:
        # Find the index of the movie
        indexs = new[new['title'].str.lower().str.strip() == movie_lower].index[0]
        distances = similarity[indexs]
        movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

        # Check if there are movies in the list
        if movie_list:
            for i in movie_list:
                print(new.iloc[i[0]].title)
        else:
            print("Not enough similar movies found for recommendations.")
    else:
        print(f"Movie '{movie}' not found in the DataFrame.")
```

```python
recommend('Avatar')
```
